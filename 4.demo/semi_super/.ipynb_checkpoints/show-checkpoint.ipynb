```python
import joblib 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def show(log_path):
    log= joblib.load(log_path)
    num = log['num_list'] 
    auc = pd.DataFrame(log['auc_list'],columns=['auc']).values
    auc_ewm = pd.DataFrame(log['auc_list'],columns=['auc']).ewm(alpha=0.2).mean().values
    f = np.polyfit(num, auc, 3)
    curve = np.polyval(f,num)
    print('random_seed:',log['random_seed'])
    print('step:',num[1])
    
    plt.plot(num,auc_ewm)
    plt.plot(num,curve,'r')
```

```python
show('log0')
show('log5')
show('log4')
show('./his_log/log2')
```
